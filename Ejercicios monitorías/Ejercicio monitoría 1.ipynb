```python
# 1. Crear una lista de 10 números y calcular la suma, promedio y varianza usando sum, len y numpy
import numpy as np

numeros = [2, 5, 8, 12, 15, 7, 9, 10, 27, 6]
# con sum y len
suma = sum(numeros)
promedio = suma / len(numeros)
print(f"Promedio con sum y len: {promedio}")

#con numpy
array_numeros = np.array(numeros)
promedio2 = np.mean(array_numeros)
varianza = np.var(array_numeros)
print(f"Promedio usando numpy: {promedio2}")
print(f"Varianza usando numpy: {varianza}")


# 2. Convertir la lista a numpy.array y multiplicarla por 3.
array_multiplicado = array_numeros * 3
print(f"Lista original: {numeros}")
print(f"Array multiplicado por 3: {array_multiplicado}")


# 3. Crear un DataFrame con pandas, calcular promedio y graficar.
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Materia': ['Física mecánica', 'Teoría de la computación', 'Banca central'],
    'Créditos': [3, 2, 3],
    'Nota estimada': [4.2, 4.8, 4.5]
}

dataframe = pd.DataFrame(data)

# Calcular el promedio ponderado
dataframe['Ponderado'] = dataframe['Créditos'] * dataframe['Nota estimada']
promedio_ponderado = dataframe['Ponderado'].sum() / dataframe['Créditos'].sum()
print(f"\nDataFrame:\n{dataframe}")
print(f"\nEl promedio ponderado de notas es: {promedio_ponderado:.2f}")

# Graficar un bar chart
plt.bar(dataframe['Materia'], dataframe['Nota estimada'], color=['skyblue', 'pink', 'salmon'])
plt.title('Notas estimadas por materia')
plt.xlabel('Materia')
plt.ylabel('Nota estimada')
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--')
plt.show()
```
